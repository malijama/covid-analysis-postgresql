```python
import pandas as pd
from sqlalchemy import create_engine, text

url = "https://raw.githubusercontent.com/owid/covid-19-data/master/public/data/owid-covid-data.csv"
df = pd.read_csv(url)

colonnes = [
    "iso_code",
    "continent",
    "location",
    "date",
    "total_cases",
    "new_cases",
    "total_deaths",
    "new_deaths",
    "total_vaccinations",
    "people_vaccinated_per_hundred",
    "population",
]
df_clean = df[colonnes].copy()
df_clean["date"] = pd.to_datetime(df_clean["date"])

PASSWORD = "iverson11"

engine = create_engine(
    f"postgresql+psycopg2://postgres:{PASSWORD}@127.0.0.1:5432/postgres"
)

with engine.connect() as conn:
    version = conn.execute(text("SELECT version();")).scalar()
    print("✅ Connexion PostgreSQL OK")
    print(version)

df_clean.to_sql(
    "covid_raw",
    engine,
    if_exists="replace",   # écrase complètement la table existante
    index=False,
    chunksize=10000
)

print("✅ Import terminé :", len(df_clean), "lignes chargées dans la table covid_raw")
```
